```python
import pandas as pd
```

```python
vocab_dir = "https://raw.githubusercontent.com/spdkh/DataMining-Course-8118/master/Week%202/C1-data/vocab.txt"
vocab = pd.read_table(vocab_dir, header=None, index_col=0)

vocab
```

```python
def fp_grow(topic_id, vocab, support):
    topic_id -= 1
    topic_dir = 'https://raw.githubusercontent.com/spdkh/DataMining-Course-8118/master/Week%202/C1-data/topic-' + str(topic_id + 1) + '.txt'
    topic = pd.read_csv(topic_dir, header=None, sep=' ', names=range(30))
    print('\nTopic', topic_id + 1, ':\n', topic.iloc[:, :5])

    vocab['counter_' + str(topic_id + 1)] = 0

    for row in topic.index:
        ind = set(filter(lambda x: x == x, topic.iloc[row, :]))
        vocab.loc[ind, 'counter_' + str(topic_id + 1)] += 1

    freq_support = vocab.sort_values(by='counter_' + str(topic_id + 1), ascending=False)
    freq_support = freq_support[freq_support.iloc[:, topic_id + 1] >= support]
    return freq_support
```

```python
support = 400

freq_supports = list()
for topic_id in range(1, 5):
    freq_supports.append(fp_grow(topic_id, vocab, support))

print('\nfrequent supports for all topic indices:\n')
freq_supports
```
